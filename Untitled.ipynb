# Exploratory Data Analysis

## Gender

```python
df.columns
```

```python
ax = sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sale_gender=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data=sale_gender)
```

From the above graphs we can see that most of buyers are females and the amount spend by females are more as compared to males.

# Age

```python
ax = sns.countplot(data=df,x='Age Group',hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

From the above graph we can see that most of the buyers are of the age group between 26-35 years female.

```python
sale_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sale_age)    
```

From the above graph we can see that most of the buyers are of the age group between 26-35 years.

# State

```python
df.columns
```

```python
# Total Numbers Of Orders From Top 10 States
sales_state=df.groupby('State',as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
# plt.figure(figsize=(18,10))
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

From the above graphs we can see that these are the top 10 states which Orders the most.

```python
# Total amount/sales from top 10 states
sales_state=df.groupby('State',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State',y='Amount',data=sales_state)
```

From the above graphs we can see that unexpectedly most of the orders are from Uttar Pradesh, Maharastra, Karnataka.
But from the last we can see that Bihar Expenditure is more than Kerala which was there in the top 10 list of states which
orders most.

# Marital Status

```python
sns.set(rc={'figure.figsize':(6,5)})
ax = sns.countplot(x='Marital_Status',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_married=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=sales_married)    
```

From the above graphs we can see that most of the buyers are married(Women) and they have high purchasing power.

# Occupation

```python
sns.set(rc={'figure.figsize':(20,8)})
ax = sns.countplot(data=df, x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Occupation',y='Amount',data=sales_occupation)    
```

From the above graphs we can see that most of the buyers are working in IT Sector, Healthcare, Aviation.

# Product Category

```python
sns.set(rc={'figure.figsize':(25,8)})
ax = sns.countplot(data=df, x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.barplot(x='Product_Category',y='Amount',data=sales_occupation)    
```

From the above graphs we can see that most of the sold products are from Food, Clothing and Electronics category.

```python
sales_productid=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(data=sales_productid,x='Product_ID',y='Orders')
```
